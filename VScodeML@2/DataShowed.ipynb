```python
import yfinance as yf
import matplotlib.pyplot as plt

dataShow = yf.Ticker("AAPL")
dataShow.actions
```

```python

actions = dataShow.actions

# Plot the actions data
fig, ax = plt.subplots(figsize=(16, 10))
actions.plot(ax=ax, kind='bar', title='AAPL Dividends and Stock Splits')
ax.set_xlabel('Date')
ax.set_ylabel('Value')
plt.show()
```
